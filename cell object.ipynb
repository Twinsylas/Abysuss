```python
from bisect import bisect
def findRadius(cluster,heaters):
    heaters.sort()
    ans=[]
    h_idx=[]
    for h in cluster:
        hi=bisect(heaters,h[1])
        left=heaters[hi-1] if hi-1>=0 else float('-inf')
        right=heaters[hi] if hi < len(heaters) else float('inf')
        if h[1]-left< right-h[1]:
            ans.append(h[1]-left)
            h_idx.append(hi-1)
        else:
            ans.append(right-h[1])
            h_idx.append(hi)
    return ans,h_idx
```
